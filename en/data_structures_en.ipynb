In some situations, we may want to create a copy of a list and modify it without affecting the content of the original list.

We can first make a shallow copy, which creates a new list containing references to the objects present in the original list. This can be done in the following ways:
* Using the slicing operator `[:]` (you already know it)
* Using the `.copy()` method
* Using the `copy()` method from the `copy` module

````
# Using [:]
brands_EU = ["BMW", "VW", "RENAULT"]
print(brands_EU[:])
````

````
brands_EU = ["BMW", "VW", "RENAULT"]
brands_USA = ["TESLA", "FORD"]
brands_cars = [brands_EU, brands_USA]
print(brands_cars) # List of lists
````

````
# Using the method .copy()
brands_cars_copied = brands_cars.copy()
print(brands_cars_copied)
````

````
#You can use el metodo copy del modulo copy
from copy import copy
brands_cars_copied = copy(brands_cars)

print('`brands_cars` id: ',hex(id(brands_cars)))
print('`brands_cars_copied`: ',hex(id(brands_cars_copied)))
````

Let's check whether the first element of each list refers to the same object.

````
# Run: 
print(id(brands_cars[0])) 
print(id(brands_cars_copied[0])) 
# Yes, they refer to the same object.
````

When we want to make a complete copy of a list, we need to make a deep copy. This approach recursively creates copies of each element of the object.

````
from copy import deepcopy

brands_EU = ["BMW", "VW", "RENAULT"]
brands_USA = ["TESLA", "FORD"]
brands_cars = [brands_EU, brands_USA]

brands_cars_deep_copied = deepcopy(brands_cars)

# Check el id de las listas
print('`brands_cars` id: ',hex(id(brands_cars)))
print('`brands_cars_deep_copied`: ',hex(id(brands_cars_deep_copied)))

# Check el id del primer objeto en las listas.

print('`brands_cars[0]` id: ',hex(id(brands_cars[0])))
print('`brands_cars_deep_copied[0]` id: ',hex(id(brands_cars_deep_copied[0])))
````

# Iterations (for loops)
As mentioned earlier, a list is a sequence of objects in a particular order. In Python, we commonly want to go through each element of this list and execute tasks. Traversing a list can be done using different constructs, the most common of which are `for` loops and list comprehensions.

#### Using `for` loops
The keyword `for` can be used to iterate over a sequence a fixed number of times. This keyword must always be used together with an iterable object. `for` iterates over the members of the sequence __in order__, executing the code block each time. The keyword `for` is used together with the keyword `in`.

````
brands_EU = ["BMW", "VW", "RENAULT","CITROEN","FIAT"]
# The pattern es el siguiente `for` <variable> `in` <secuencia>: <bloque de codigo>
for brand in brands_EU:
    print(brand)
````

````
# Podemos iterar through cualquier secuencia 
# We can iterate through any sequence
for letter in brands_EU[2]:
    print(letter)
````

We can build nested iterations if necessary. We need to repeat the pattern `for` <variable> `in` <sequence>: <code block>, paying attention to indentation, which defines the scope of the `for` loop.

````
for brand in brands_EU: # 5 elements
    my_variable = 'happy'
    print(f"This is the brand: {brand}")
    # Repeat the pattern
    for letter in brand: 
        print(f"This is a letter: {letter}")
        ...
````

In some situations, we also want to know the index location of the current variable. For this, we can use the built-in function `enumerate()`, which takes an iterable and returns an iterator that produces a tuple of two elements when requested.

````
brands_EU = ["BMW", "VW", "RENAULT","CITROEN","FIAT"]
# The pattern es `for` <variable> `in` <sequence>: <code block>
for index, brand in enumerate(brands_EU): # enumerate() devuelve dos objetos: el indice y el valor.

    print(f"This is the brand: {brand} located at position: {index}")
````

<div class="alert alert-block alert-warning"><b>Warning: </b>`enumerate()` creates an iterator that returns two objects after each loop. Unlike a sequence, where all elements of the sequence are already in memory, an iterator produces values progressively, on demand.</div>

The `for` loop will iterate until the list is exhausted. In some cases, we may want to stop the iteration after a condition is met. For this, we use the keyword `break`.

````
brands_EU = ["BMW", "VW", "RENAULT","CITROEN","FIAT"]

for brand in brands_EU:

    print(f"Esta es una marca: {brand}")

    if brand == "RENAULT": # Condition introduced
        print("Condition reached; stopping the loop")
        break
    # we repeat the pattern
    for letter in brand:
        print(f"Esta es una letra: {letter}")
````

<div class="alert alert-block alert-info"><b>Tip: </b>The keyword `if` is used to introduce a condition. The pattern is `if` <expression>: <code block>. The <code block> will be executed only if the expression is `True`.</div>

````
# Ejemplo
if 3>2:
    print("Todo en orden")
````

Similarly to the previous case, we can skip the code block if a condition is met without stopping the whole iteration. For this, we use the keyword `continue`.


````
brands_EU = ["BMW", "VW", "RENAULT", "CITROEN", "FIAT"]

for brand in brands_EU:

    if brand == "RENAULT":  # Condición introducida
        print("Condition met; skipping this brand")
        continue
    print(f"This is the brand: {brand}")
````

<div class="alert alert-block alert-info"><b>Tip: </b>The structure of the `for` loop exists in many programming languages with different syntax. In other programming languages, you may need to define expressions to control the value of the next cycle or an expression to determine whether the cycle has ended. In Python, this is controlled by building the appropriate sequence.</div>

#### Using list comprehensions to iterate
List comprehensions are a particular Python construct and allow us to iterate through lists with a single line of code. The main difference compared with `for` loops is that list comprehensions are designed to create a list through iteration over another list. 

A list comprehension combines the `for` loop and the creation of new elements in a single line. This is an advanced way to build lists, but you will probably find it in many Python files.


````
# Bucle for y methods incorporados para modificar listas
brands_EU = ["BMW", "VW", "RENAULT", "CITROEN", "FIAT"]
selected = []

for brand in brands_EU:
    if 'W' in brand:  # Condición introducida
        selected.append(brand)

print(f"This is the new list: {selected}")
````

````
# Using list comprehensions

# El patrón es [<variable_1> for <variable_1> in <una secuencia> <condición>]
selected = [brand for brand in brands_EU if 'R' in brand]

print(f"This is the new list: {selected}")

````

````

# Using list comprehensions

# El patrón es [<variable_1> for <variable_1> in <una secuencia> <condición>]
selected = [brand for brand in brands_EU if 'R' in brand]

# We can also make nested iterations con comprensión de listas
brands_EU = ["BMW", "VW", "RENAULT"]
brands_USA = ["GMC", "TESLA", "FORD"]
all_brands = [brands_EU, brands_USA]
print(f"All brands: {all_brands}")

# Using for loops
selected = []
for brand_list in all_brands:
    for brand in brand_list:
        if "L" in brand:
            selected.append(brand)
print('Using for loops: ', selected)

# Using list comprehensions
selected_1 = [brand for brand_list in all_brands for brand in brand_list if "L" in brand]

print('Using list comprehensions: ', selected_1)
````

## Dictionaries  

Dictionaries are data structures that allow us to connect pieces of related information. Anything that can be paired can be stored as a dictionary. Dictionaries are important data structures in Python because they can store almost unlimited nested data.  



````
# A dictionary is a collection de key-value pairs.  
# Each key is unique, y el valor puede ser de cualquier tipo de dato (int, string, lista, etc.)  

# Example: create un diccionario  
persona = {  
    "nombre": "John",  
    "edad": 30,  
    "ciudad": "Nueva York",  
}  

# Print el diccionario  
print(persona)  
````

````
# We can access a los valores en un diccionario usando la clave  
# En este caso, la clave es 'nombre'  
print(persona["nombre"])  
````

````
# We can also use el method `get()` para acceder a los valores  
print(persona.get("nombre"))  
````

````
# If the key does not exist, `get()` devuelve None en lugar de generar un error  
type(persona.get("país"))
````

````
# Add a new pair clave-valor  
persona["comida_favorita"] = "pizza"  
print(persona)  # Ahora el diccionario incluye la clave 'país'  
````

````
# Update a value existente  
persona["edad"] = 42  
print(persona)  # El valor de 'edad' se actualiza a 42  
````

````
# Use the keyword `del`  
del persona["ciudad"]  
print(persona)  # 'ciudad' se elimina del diccionario
````

````
# Usar `pop()` para remove and return el valor  
edad = persona.pop("edad")  
print(persona)  # 'edad' se elimina  
print(f"Edad eliminada: {edad}")  # Salida: Edad eliminada: 31  
````

````
# Check whether a key exists en un diccionario  

# Use the keyword `in`  
print("nombre" in persona)  # True  
print("ciudad" in persona)  # False 
````

````
# Traverse a dictionary  

# Iterate through de las claves  
for clave in persona:  
    print(clave) 
````

````
# Iterate through de los valores  
for valor in persona.values():  
    print(valor)  # Imprime todos los valores 
````

````
# Iterate through de key-value pairs  
for clave, valor in persona.items():  
    print(f"La respuesta es: {clave}: {valor}")  # Imprime key-value pairs  
````

````
# También podemos aplicar una dictionary comprehension  

# Crear un nuevo diccionario donde las claves son números y los valores son sus cuadrados  
cuadrados = {(x): f"mi número es {x}" for x in range(5)}  
print(cuadrados)  
````

````
# Nested dictionaries (diccionarios dentro de diccionarios)  

# Un diccionario que contiene otro diccionario como valor  
diccionario_anidado = {  
    "persona1": {"nombre": "Alice", "edad": 25},  
    "persona2": {"nombre": "Bob", "edad": 28}  
}  

# Acceder a un diccionario anidado  
# print(diccionario_anidado["persona1"]["nombre"])  

for clave, valor in diccionario_anidado.items():  
    print(f"El nombre de nuestro cliente es {valor['nombre']}")  
````

````

# Dictionary methods  

# `keys()` devuelve todas las claves como un objeto dict_keys  
print(persona.keys())

# `values()` devuelve todos los valores como un objeto dict_values  
print(persona.values())

# `items()` devuelve todos los key-value pairs como un objeto dict_items  
print(persona.items())  

# Use `update()` para fusionar otro diccionario en el actual  
persona.update({"género": "Masculino", "ciudad": "Nueva York"})  
print(persona)
````

````
# Usar el unpacking operator `**` para fusionar diccionarios  
nueva_informacion = {1: ["Leer", "Viajar"]}  
persona = {  
    "nombre": "John",  
    "edad": 30,  
    "ciudad": "Nueva York",  
}  
diccionario_fusionado = {**persona, **nueva_informacion}  
print(diccionario_fusionado)
````

## Tuples
A tuple is an ordered and immutable collection of elements. Tuples are similar to lists, but unlike lists, they cannot be modified once created.


````
# Example: create una tupla
mi_tupla = (1, 2, 3, 4)
print(mi_tupla)
````

````
# Access elements en una tupla

# Puedes acceder a los elementos de una tupla por índice (indexación basada en 0)
print(mi_tupla[0])  
print(mi_tupla[2])  

# La indexación negativa comienza desde el final
print(mi_tupla[-1]) 
````

````
# Slice a tuple

# Rebanar devuelve una nueva tupla con un subconjunto de los elementos
print(mi_tupla[1:3])  

# Rebanar con indexación negativa
print(mi_tupla[-3:-1])
````

````
# Tuples are immutable, lo que significa que sus elementos no se pueden cambiar después de su creación

# Trying to modify un elemento will result in an error
mi_tupla[0] = 10  
````

````
# Concatenate and repeat tuplas

# Concatenar dos tuplas
tupla1 = (1, 2)
tupla2 = (3, 4)
tupla_combinada = tupla1 + tupla2
print(tupla_combinada)  

# Repetir una tupla
tupla_repetida = (5, 6) * 3
print(tupla_repetida)  
````

````
# Check the length de una tupla

# Usar la función len() para obtener el número de elementos en una tupla
print(len(mi_tupla))  
````

# Functions

The concept of functions in Python is analogous to the concept of functions in mathematics.

 $$ z = f(x,y)$$

In Python, functions are more versatile and generic than mathematical functions. In fact, they contain reusable code blocks that can perform a specific task.
They help organize code, reduce redundancy, and improve modularity.

Basic syntax
```python
def function_name(<your-parameters>):
    <your code block here>
    return <an output value>      # Optional return statement
```
- **`def`**: Keyword used to define a function
- **<your-parameters>**: Function inputs (optional)
- **`return`**: Returns a value (optional; default is `None`)

````
# A basic function necesita usar indentación para indicar el alcance
def saludar():
    return "¡Hola, Mundo!"

mensaje = saludar()  # Llamada a la función
print(mensaje)
````

````
def sumar(a, b):
    """Devuelve la suma de dos números""" 
    return a + b

resultado = sumar(3, 5)
print(f"3 + 5 = {resultado}")
````

````
sumar(2,1)
````

````
# Podemos crear default parameters usando el operador =
def potencia(base, exponente=2):
    """Devuelve la base elevada al exponente (por defecto: al cuadrado)"""
    return base ** exponente
print(potencia(2))
````

````
print(potencia(2,5))
````

````
# Arguments are passed a cada parámetro correspondiente usando el operador de asignación
def info_persona(nombre, edad, ciudad):
    print(f"{nombre}, {edad} años, vive en {ciudad}")

info_persona(edad=30, ciudad="París", nombre="Alicia")  # El orden no importa
````

In Python, we can receive arguments flexibly using `*args` and `**kwargs`.


````
def imprimir_args(*args, **kwargs):
    print("Positional arguments:", args)
    print("Keyword arguments:", kwargs)

imprimir_args(1, 'a', nombre="Juan", edad=25, a=1, b=3)
````

````
def validar(*args, **pizza):
    if 'cuenta' in pizza:
        print(f'Access granted to: {pizza["cuenta"]}')
    else:
        print("Invalid")

validar(1, nombre="Juan", edad=25, a=1, b=3, cuenta='Gustavo', tamaño='L')
````

A function can return one or more values when `return` is defined; otherwise, it returns `None`.

About arguments and parameters:

- **Required parameters**: Must be passed when calling the function
- **Default parameters**: Have default values if not provided
- **Keyword arguments**: Specified by the parameter name
- **Variable-length arguments**: `*args` (tuples) and `**kwargs` (dictionaries)

# `If` conditionals
The `if` statement executes code blocks based on Boolean conditions.

**Syntax:**
```python
if condition:
    # indented code block
```

**Key rules:**
- Colon `:` required after the condition
- Code blocks are defined by indentation (4 spaces are recommended)
- Comparison operators: `>`, `<`, `==`, `!=`, `>=`, `<=`

````
temperatura = 30
if temperatura > 25:
    print("¡It is a hot day!")
    print("¡Stay hydrated!")
````

## If-Else statement
Provides alternative execution paths:
```python
if condition:
    # true case
else:
    # false case
```

````
edad = 12
if edad >= 18:
    print("You are eligible to vote")
else:
    print("You are not yet eligible to vote")
    años_faltantes = 18 - edad
    print(f"Come back in {años_faltantes} año{'s' if años_faltantes > 1 else ''}")
````

## Elif statements (Else If)
Handles multiple exclusive conditions:
```python
if cond1:
    ...
elif cond2:
    ...
else:
    ...
```

````
puntuación = 82
if puntuación >= 90:
    calificación = "A"
elif puntuación >= 80:
    calificación = "B"
elif puntuación >= 70:
    calificación = "C"
elif puntuación >= 60:
    calificación = "D"
else:
    calificación = "F"
print(f"Score: {puntuación} -> Grade: {calificación}")
````

## Truthiness and Boolean evaluation
Python evaluates non-Boolean values in conditions:
- **Falsy:** `None`, `0`, empty collections (`""`, `[]`, `{}`, etc.), False
- **Truthy:** All other values, True

````
nombre_usuario = 4
if nombre_usuario:
    print(f"Welcome {nombre_usuario}")
else:
    print("Guest user")
````

## Logical operators
Combine conditions with:
- `and`: Both must be true
- `or`: At least one must be true
- `not`: Inverts the Boolean value

````
edad = 45
tiene_licencia = False
if (edad >= 18) or (tiene_licencia):
    print("You can drive a car")
else:
    print("Driving is not allowed")
````

## Nested if statements
`if` statements can contain other `if` statements.

````
saldo_cuenta = 11500
es_premium = 1
if saldo_cuenta > 0:
    print("Active account")
    if es_premium:
        print("Premium benefits activated")
        if es_premium and saldo_cuenta > 2000:
            print("Congratulations, you won a prize")
    else:
        print("Standard account")
else:
    print("Suspended account")
````

## Ternary operator
Compact conditional expression:
```python
result = true_value if condition else false_value
```

````
costo = 99
var = 24 if costo < 15 else 56
print(var)
````

````
temperatura = 28
estado = "Caluroso" if temperatura > 25 else "Fresco"
print(f"Status: {estado}")
````

## Practical example: User access control
Real-world implementation combining multiple concepts

````
usuario = {
    "nombre_usuario": "admin",
    "está_activo": True,
    "intentos_de_inicio": 2,
    "rol": "admin"
}

if usuario["está_activo"]:
    if usuario["intentos_de_inicio"] < 3:
        if usuario["rol"] == "admin":
            print("Administrator panel")
        elif usuario["rol"] == "editor":
            print("Editor panel")
        else:
            print("User profile")
    else:
        print("Locked account")
else:
    print("Deactivated account")

````
