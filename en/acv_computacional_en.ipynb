In LCA, we seek to understand the environmental impacts across the entire life cycle of a product or service. However, to estimate these impacts, it is first important to understand the system in which the product or service operates. This system is represented by a supply-chain model that contains the multiple stages of the product life cycle.

Imagine that we want to calculate the life-cycle impact of producing a bicycle in an extremely simplified system. In this case, we can interpret the graph (a collection of nodes and edges) as follows:

> *To produce 1 bicycle, 2.5 kg of carbon fiber are needed. In turn, to produce 1 kg of carbon fiber, 237 MJ of natural gas are required, and 26.6 kg of CO2 are emitted during the process.*

![image info](./media/simple-graph.png)

In the figure, the circular nodes represent activities that occur in the technosphere, while the rectangular nodes represent elements of the biosphere. As mentioned above, the first step is to determine how many units of each technosphere product will be produced as a result of the bicycle demand, and consequently how many emissions are generated in the process. This information then allows us to estimate the total impacts over the whole life cycle.

The most common way to approach this problem mathematically is through a linear algebra problem, as clearly presented and discussed in [this book](https://link.springer.com/book/10.1007/978-94-015-9900-9). In this approach, the technosphere is represented as a matrix in which each column contains the “recipe” for producing one unit of a given product (matrix $A_{products \times activities}$). The biosphere is represented as another matrix with dimensions $B_{flows \times activities}$, and the demand (functional unit) is represented as a vector $f_{products}$. In addition, a matrix $Q_{impacts \times flows}$ is introduced to convert flows into impacts through characterization factors.

In this representation, positive values indicate “output”, while negative values indicate “consumption”.

$$ A = \begin{bmatrix} 1 & 0 & 0  \\ -2.5 & 1 & 0 \\ 0 & -237 & 1  \end{bmatrix}$$

$$ B = \begin{bmatrix} 0 & 26.6 & 0  \end{bmatrix}$$

$$ y = \begin{bmatrix} 0 \\ 1 \\ 0 \end{bmatrix}$$

$$ Q = \begin{bmatrix} 1 \end{bmatrix}$$

The problem is now to obtain a supply vector $s_{products}$, which represents all flows in the technosphere resulting from the demand vector $f$. Fortunately, this problem was solved many years ago in the field of Input-Output analysis, and we only need to solve the following equation:

$$ s = A^{-1}f $$

Once $s$ has been obtained, we can calculate all environmental flows using $g = Bs$, and all impacts by solving $h = Qg$.

This system of equations can be solved easily using any package or library that can invert the matrix $A$. To demonstrate this example, we will use the [NumPy](https://numpy.org/) package, which allows us to manipulate multidimensional arrays quickly and efficiently. As you will see later, Brightway has been built using NumPy as its calculation engine.


```
# We import the NumPy and SciPy libraries.
# NOTE: We are not using Brightway yet; this is only an example.
import numpy as np
from rich import print  # Only to improve the appearance of printed outputs

A = np.array([[1, 0, 0], [-2.5, 1, 0], [0, -237, 1]])
B = np.array([[0, 26.6, 0]])
f = np.array([1, 0, 0])
Q = np.array([1])

# Check the matrices and vectors
print(f"The dimensions of A are: {A.shape}")
print("A: \n", A)

print(f"The dimensions of vector f are: {f.shape}")
print("f: \n", f)
```

```

A_inv = np.linalg.inv(A)  # np.linalg.inv allows us to invert matrices
s = A_inv.dot(f)
g = B.dot(s)

print("s: ", s)
print("g: ", g)
```

With the vectors $s$ and $g$ calculated, it is possible to obtain the vector $h$, which represents the vector of environmental impacts.

We could say that this is the `final` step of the methodological workflow in life cycle assessment.


```

h = Q.dot(g)
print("h: ", h)
```
