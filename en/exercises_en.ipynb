
🚧 Life Cycle Inventory Exercises

Below you will find 6 exercises to practice building LCA inventories.
Each case presents a different product with realistic but approximate data for the analysis.


## Exercise 1: Bicycle production

    The bicycle business continues to grow, and it is time to consider increasing production. You are aware that if you want to reach more markets, you will need to provide wholesalers with more information, especially on environmental aspects.
    The engineers have suggested considering the import of Chinese aluminum, or even moving the production plants to China.
    Your task is to build two models and explore the possible differences between producing bicycles in Peru and producing them in China.
    You should consider the following:

    The estimated consumption per bicycle is:

        345 kWh of energy
        5 kg of steel
        25 kg of aluminum
        The final product weighs 24 kg
        Around 15 liters of diesel are consumed


## Exercise 2: Laptop manufacturing

    A technology company is evaluating the environmental footprint of its laptops to comply with European regulations.
    You need to compare production in two locations: a factory in Malaysia and another in Mexico.
    The company is seeking environmental certification and requires a detailed analysis of the material inventory.

    The estimated consumption per laptop is:

        180 kWh of electricity
        2.5 kg of various plastics (ABS, PC)
        0.8 kg of aluminum for the casing
        0.15 kg of copper for circuits
        0.05 kg of gold and precious metals
        The final product weighs 1.8 kg
        8 liters of water are consumed in cooling processes


## Exercise 3: Solar panel production

    A renewable energy startup wants to expand into Latin American markets and needs to evaluate
    the environmental impact of its photovoltaic solar panels.
    It is considering establishing local production in Colombia versus importing from Germany.
    The analysis should include both manufacturing and transport to the final user.

    The estimated consumption per solar panel (300W) is:

        520 kWh of energy during manufacturing
        12 kg of purified silicon
        15 kg of tempered glass
        8 kg of aluminum for the frame
        2 kg of polymers (EVA, backsheet)
        The final product weighs 22 kg
        25 liters of cleaning chemicals are consumed


## Exercise 4: Glass bottle manufacturing

    A craft beverage company wants to reduce its environmental impact and is evaluating
    a switch from plastic bottles to reusable glass bottles.
    It needs to compare the production of recycled glass bottles versus virgin glass bottles,
    considering that the bottles will have an average life cycle of 15 reuses.

    The estimated consumption per glass bottle (500 ml) is:

        0.8 kWh of energy for melting
        0.45 kg of silica sand (or equivalent recycled glass)
        0.05 kg of sodium carbonate
        0.02 kg of limestone
        The final product weighs 0.35 kg
        2 liters of water are consumed for cooling


## Exercise 5: Cotton T-shirt production

    A sustainable clothing brand wants to quantify the impact of its basic T-shirts.
    It is evaluating three options: conventional cotton from India, organic cotton from Peru,
    and a recycled cotton blend produced in Turkey.
    The analysis should include cultivation/recycling, spinning, weaving, dyeing, and garment making.

    The estimated consumption per T-shirt is:

        12 kWh of total process energy
        0.25 kg of cotton fiber
        0.03 kg of dyes and chemicals
        0.01 kg of auxiliary threads
        The final product weighs 0.18 kg
        150 liters of water are consumed in wet processes


## Exercise 6: Ceramic brick manufacturing

    A construction company is implementing sustainable construction criteria and needs to evaluate
    different types of bricks for its housing projects.
    It wants to compare traditional clay bricks with bricks containing recycled content,
    both produced in kilns using different fuels (natural gas vs. biomass).

    The estimated consumption per standard brick is:

        1.2 kWh of thermal energy for firing
        2.8 kg of clay and aggregates
        0.2 kg of fuel (natural gas or biomass)
        The final product weighs 2.5 kg
        0.8 liters of water are consumed for molding
        Approximately 1.1 kg of CO₂ is generated by combustion


For each exercise, you should:
1. Clearly define the functional unit
2. Establish the system boundaries
3. Create the input and output inventory using Brightway
4. Consider transport processes when relevant. Approximate distances by ship, airplane, or road transport can be used.
5. Perform the analysis for 2 categories of the EF v3.1 method
6. Analyze the results and propose improvement alternatives based on the results
