Example: Carrot Production in Switzerland (Conventional vs. Organic)

This is a teaching example showing how to build a Life Cycle Inventory (LCI) using the BAFU database (based on ecoinvent).
In this simplified example, we will calculate the environmental impact of 1 kg of carrots produced in Switzerland (CH) under two different agronomic systems.

You must calculate the climate change impact and compare both options.

## SYSTEM 1: CONVENTIONAL CARROT PRODUCTION (SWITZERLAND)
The estimated average yield is 50 t (50,000 kg) of carrots per hectare.
The following agronomic information was collected per hectare:
 - Synthetic fertiliser consumption: 100 kg N/ha, 50 kg P2O5/ha, 150 kg K2O/ha
 - Diesel consumption for machinery (sowing, tillage, harvesting): 300 L/ha (approx. 10,800 MJ/ha)
 - Synthetic pesticide application: 2 kg/ha

## SYSTEM 2: ORGANIC CARROT PRODUCTION (SWITZERLAND)

The yield is usually lower in organic systems, estimated at 20 t (20,000 kg) of carrots per hectare.
The following agronomic information was collected per hectare:
 - 25 t/ha of manure or organic compost is applied.
 - Diesel consumption for machinery: 350 L/ha (approx. 12,600 MJ/ha). Consumption is slightly higher due to the intensive use of machinery for mechanical weed control, given the absence of herbicides.
